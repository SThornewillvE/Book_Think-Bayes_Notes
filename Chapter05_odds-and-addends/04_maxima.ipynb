# Maxima

Now that we've thought about the distributions of sums, how about the distributions of maximums?

We can calculate it via the same other methods as before, simulation and enumeration. However this time we also have a third method, exponentiation.

If we convert a Pdf to a Cdf there is a simple and efficient algorithm for finding the Cdf for the maximum.

```python
# Import packages
from thinkbayes2 import Pmf, MakePmfFromList, Cdf
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("seaborn")
```

```python
# Define dice
class Die(Pmf):
    def __init__(self, sides):
        Pmf.__init__(self)
        for x in range(1, sides+1):
            self.Set(x, 1)
        self.Normalize()
```

Once again, we want to create 3xD6 dice. 

```python
# Create D6
d6 = Die(6)

three_exact = d6 + d6 + d6
```

Then, we want to take the maximum value of 6 values chosen at random and then convert from cdf to pmf.

```python
best_attr_cdf = three_exact.Max(6)
best_attr_pmf = best_attr_cdf.MakePmf()
```

```python
# Get Data to plot result
x = []
y = []

for hypothesis, probability in best_attr_pmf.Items():
    x.append(hypothesis)
    y.append(probability)

fix, ax = plt.subplots(figsize=(10, 6), edgecolor='k', facecolor='white')
plt.plot(x, y)
plt.xlabel("Sum of three d6")
plt.ylabel("Probability")
plt.show()
```

Using this method, we were able to find the distribution for the maximum of this kind of a distribution.
